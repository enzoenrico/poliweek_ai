```python
from spoti import (
    get_user,
    get_songs_from_artist,
    get_recommended_by_genre,
    play_song_on_device,
    get_artist_id,
    search_by_track_name,
    select_url_from_track,
    get_artist_albums,
    get_songs_from_album,
    
)

tools = [
    get_user,
    get_songs_from_artist,
    get_recommended_by_genre,
    play_song_on_device,
    get_artist_id,
    search_by_track_name,
    select_url_from_track,
    get_artist_albums,
    get_songs_from_album,
    

]

print(tools)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a cool DJ, whose purpose is finding the best music, artists, and tailoring the most amazing playlists for your friends"),
    ("user", "{input}"),
    MessagesPlaceholder(variable_name="agent_scratchpad"),
])
```

```python
llm_tools = llm.bind_tools(tools=tools)
```

```python
from langchain.agents.format_scratchpad.openai_tools import format_to_openai_tool_messages

from langchain.agents.output_parsers.openai_tools import OpenAIToolsAgentOutputParser

agent = (
    {
        "input": lambda x: x['input'],
        "agent_scratchpad": lambda x: format_to_openai_tool_messages(x['intermediate_steps']),
    }
   | prompt
   | llm_tools
   | OpenAIToolsAgentOutputParser() 
)
```

```python
from langchain.agents import AgentExecutor

agent_exec = AgentExecutor(agent=agent, tools=tools, verbose=True)
```

```python
list(agent_exec.stream({'input': "play me a song from the latest gojira album"}))
```
